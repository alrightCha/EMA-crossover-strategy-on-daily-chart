```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pot

df = pd.read_csv('data.csv')

df # before changing the base index
```

```python
df = df.set_index(pd.DatetimeIndex(df['Date'].values)) #setting index as date as we make change in order to the date
df #after chaging the base index
```

```python
#getting the 20 and 50 daily EMA with the closing price 
fastEMA = df['Close'].ewm(span=20, adjust=False, min_periods=12).mean()
slowEMA = df['Close'].ewm(span=50, adjust=False, min_periods=12).mean()

#plotting each ema to the corresponding one to be able to view them 

df['EMA_20'] = df.index.map(fastEMA)
df['EMA_50'] = df.index.map(slowEMA)

# View our data
pd.set_option("display.max_columns", None)
df

# add columns to table

#Assigning the closing price to each day in order to get the bitcoin price action over the years

column_list =[ 'EMA_20', 'EMA_50', 'Close']

#plotting the given data in chart
df[column_list].plot(figsize=(20,10)) #determining figure size

pot.title('EMA-Crossover Bitcoin Trading Strategy') #title for the Chart
pot.ylabel('Price in dollars') # Assigning the y-axis to dollars
pot.xlabel('Date in years') # Assigning the x-axis to the timeframe
pot.yticks(rotation = 15) #rotation of price by 15 degree 
pot.show() #Display the  Graph

new_df = df[45-1:]
new_df
```

```python
def all_signals(data): 
    long_signals = [] #these will be the positions where we are bullish
    short_signals = [] #these will be the positions where we are bearish
    for i in range (len(data['Close'])):  
        if i== 0: 
            pass
        if (data['EMA_20'][i] > data['EMA_50'][i]) and (data['EMA_20'][i-1] <= data['EMA_50'][i-1]):
            long_signals.append(data['Close'][i])
            short_signals.append(np.nan);
        elif (data['EMA_20'][i] < data['EMA_50'][i]) and (data['EMA_20'][i-1] >= data['EMA_50'][i-1]):
            short_signals.append(data['Close'][i])
            long_signals.append(np.nan);
        else:
                long_signals.append(np.nan);
                short_signals.append(np.nan);
    return long_signals, short_signals

# create two new columns with the signals we should receive, seperately:

new_df['long_signals'] = all_signals(new_df)[0]
new_df['short_signals'] = all_signals(new_df)[1]

#plotting all the data

fig = pot.figure(figsize=(30,15)) #size of graph
#add plot
gray = fig.add_subplot(1,1,1)
#get index values of data frame
x_axis = new_df.index
#plot closing price average
gray.plot(x_axis, new_df['Close'], color ='black', lw =3, label='Close Price', alpha= 0.1)
gray.plot(x_axis, new_df['EMA_20'], color ='blue', lw =3, label='fast EMA', alpha= 0.1)
gray.plot(x_axis, new_df['EMA_50'], color ='red', lw =3, label='slow EMA', alpha= 0.1)
gray.scatter(x_axis, new_df['long_signals'], color ='green', lw=3 , label='Long position', marker='^', alpha= 1)
gray.scatter(x_axis, new_df['short_signals'], color ='red', lw=3 , label='Short position', marker='v', alpha= 1)

gray.set_title('EMA crossover strategy (20/50)')
gray.set_xlabel("time")
gray.set_ylabel('price in $')
pot.yticks(rotation = 15) #rotation of price by 15 degree 
gray.legend()  #showing the legend
pot.show() #displaying the graph
pd.set_option("display.max_rows", 100)


```

```python
longs = []
shorts = []
open_pos = False
startingBag = 100 #starting capital in dollars

for i in range(len(new_df)):
    if (i==0):
        pass
    if (new_df['EMA_20'][i] > new_df['EMA_50'][i])  and (new_df['EMA_20'][i-1] <= new_df['EMA_50'][i-1]): 
        longs.append(i)
        open_pos = True
    elif (new_df['EMA_20'][i] < new_df['EMA_50'][i]) and (new_df['EMA_20'][i-1] >= new_df['EMA_50'][i-1]):
        shorts.append(i)
        open_pos = False
```

```python
total = pd.concat([df.iloc[longs].Close, df.iloc[shorts].Close], axis = 1)
total.columns = ['Buys', 'Sells']
total
```

```python
totalprofit = total.shift(-1).Sells - total.Buys
totalprofit
```

```python
relprofit = (total.shift(-1).Sells - total.Buys)/total.Buys
relprofit
```
